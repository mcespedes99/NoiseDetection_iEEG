```python
import scipy.fft as scifft
srate=1024
frex = 480
nCycles = 35

time = np.arange(
        -1, 1 + 1 / srate, 1 / srate
    ) 
nKern = len(time)

nData = 3*srate

nConv = nKern + nData - 1

s = nCycles / (2 * np.pi * frex)

cmw = np.multiply(
    np.exp(2 * 1j * np.pi * frex * time),
    np.exp(-(time**2) / (2 * s**2)),
)

pnts = len(time);

mwX = abs(scifft.fft(cmw)/pnts);
hz  = np.linspace(0,srate,pnts);
```
